```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from tqdm import tqdm
from statsmodels.discrete.discrete_model import NegativeBinomial as NegativeBinomialP

# --- Setup ---
target_col = "x_202304"
imd_rank_col = "IMD rank (1 most deprived 10% of LSOAs)"
imd_score_col = "IMD score"
all_lagged_cols = sorted(
    [col for col in df3.columns if col.startswith("x_") and col != target_col],
    key=lambda x: int(x.split("_")[1]),
    reverse=True
)

# --- Regression settings ---
min_rows_required = 20  # skip if fewer rows
test_size = 0.3         # 70/30 split

# --- Function to run one regression ---
def run_regression(y, X, test_size=0.3):
    try:
        # Convert all data to numeric
        X = X.apply(pd.to_numeric, errors="coerce")
        y = pd.to_numeric(y, errors="coerce")

        # Combine and drop NA
        data = pd.concat([X, y.rename("y")], axis=1).dropna()

        if data.shape[0] < min_rows_required:
            return np.nan, np.nan, "Too few valid rows"

        X_clean = data.drop(columns="y")
        y_clean = data["y"]

        # Train/test split
        X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=test_size, random_state=42)

        # Fit Negative Binomial model with estimated alpha
        X_train_const = sm.add_constant(X_train)
        model = NegativeBinomialP(y_train, X_train_const).fit(disp=False)

        # Predict and evaluate
        X_test_const = sm.add_constant(X_test)
        y_pred = model.predict(X_test_const)

        r2 = r2_score(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)

        return r2, mse, None

    except Exception as e:
        return np.nan, np.nan, str(e)

# --- Run all regressions ---
results = []

for base_var in [imd_rank_col, imd_score_col]:
    for lag in range(len(all_lagged_cols)):
        predictors = [base_var] + all_lagged_cols[:lag+1]
        X = df3[predictors]
        y = df3[target_col]
        r2, mse, error = run_regression(y, X, test_size=test_size)

        results.append({
            "BaseVar": base_var,
            "LagsIncluded": lag,
            "Predictors": predictors,
            "R2": r2,
            "MSE": mse,
            "Error": error
        })

# --- Convert to DataFrame and filter valid results ---
results_df = pd.DataFrame(results)
valid_results = results_df.dropna(subset=["R2", "MSE"])

# --- Get best models ---
best_r2_model = valid_results.loc[valid_results["R2"].idxmax()] if not valid_results.empty else None
best_mse_model = valid_results.loc[valid_results["MSE"].idxmin()] if not valid_results.empty else None

# --- Print summary ---
print(f"\n✅ Valid regressions: {len(valid_results)} / {len(results_df)}")
print(f"❌ Failed regressions: {results_df['Error'].notna().sum()}")

if best_r2_model is not None:
    print("\n📈 Best R² model:")
    print(best_r2_model[["BaseVar", "LagsIncluded", "R2", "MSE", "Predictors"]])

if best_mse_model is not None:
    print("\n📉 Best MSE model:")
    print(best_mse_model[["BaseVar", "LagsIncluded", "R2", "MSE", "Predictors"]])
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.discrete.discrete_model import NegativeBinomial
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# --- Setup ---
target_col = "x_202304"
base_var = "IMD rank (1 most deprived 10% of LSOAs)"
num_lags = 19

# Get lagged columns sorted (descending)
all_lagged_cols = sorted(
    [col for col in df3.columns if col.startswith("x_") and col != target_col],
    key=lambda x: int(x.split("_")[1]),
    reverse=True
)

selected_lags = all_lagged_cols[:num_lags]
predictors = [base_var] + selected_lags

# Prepare modeling data
df_model = df3[[target_col] + predictors].copy()
df_model = df_model.apply(pd.to_numeric, errors='coerce').dropna()

X = df_model[predictors]
y = df_model[target_col]

# Add constant
X_const = sm.add_constant(X)

# --- Train/Test Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_const, y, test_size=0.3, random_state=42
)

# --- Fit Negative Binomial with estimated alpha ---
model = NegativeBinomial(y_train, X_train)
result = model.fit(disp=False)  # suppress output

# --- Predict on Test Set ---
y_pred = result.predict(X_test)

# --- Evaluation ---
mse = mean_squared_error(y_test, y_pred)
ss_res = np.sum((y_test - y_pred) ** 2)
ss_tot = np.sum((y_test - np.mean(y_test)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# --- Output ---
print(result.summary())
print(f"\nTest set MSE: {mse:.6f}")
print(f"Test set R² (OLS formula): {r2:.6f}")
print(f"Estimated alpha (dispersion): {result.params[-1]:.6f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from tqdm import tqdm
from statsmodels.discrete.discrete_model import NegativeBinomial as NegativeBinomialP

# --- Settings ---
target_col = "x_202304"
base_predictors = ["imd_score_no_crime", "Population Density"]

# Sort lagged columns descending by date (most recent first)
all_lagged_cols = sorted(
    [col for col in df4.columns if col.startswith("x_") and col != target_col],
    key=lambda x: int(x.split("_")[1]),
    reverse=True
)

min_rows_required = 20
test_size = 0.3

# --- Function to run a regression ---
def run_regression(y, X, test_size=0.3):
    try:
        X = X.apply(pd.to_numeric, errors="coerce")
        y = pd.to_numeric(y, errors="coerce")
        data = pd.concat([X, y.rename("y")], axis=1).dropna()

        if data.shape[0] < min_rows_required:
            return np.nan, np.nan, "Too few valid rows"

        X_clean = data.drop(columns="y")
        y_clean = data["y"]

        X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=test_size, random_state=42)

        X_train_const = sm.add_constant(X_train)
        poisson_model = sm.GLM(y_train, X_train_const, family=sm.families.Poisson()).fit()
        pearson_chi2 = sum(((y_train - poisson_model.mu) ** 2) / poisson_model.mu)
        alpha_est = pearson_chi2 / poisson_model.df_resid
        nb_fam = sm.families.NegativeBinomial(alpha=alpha_est)
        model = sm.GLM(y_train, X_train_const, family=nb_fam).fit()


        X_test_const = sm.add_constant(X_test)
        y_pred = model.predict(X_test_const)

        r2 = r2_score(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)

        return r2, mse, None

    except Exception as e:
        return np.nan, np.nan, str(e)

# --- Run regressions for 0 to 158 lags ---
results = []

for num_lags in tqdm(range(0, len(all_lagged_cols) + 1)):
    selected_lags = all_lagged_cols[:num_lags]  # 0 lags = no lagged predictors
    predictors = base_predictors + selected_lags

    X = df4[predictors]
    y = df4[target_col]

    r2, mse, error = run_regression(y, X, test_size=test_size)

    results.append({
        "LagsIncluded": num_lags,
        "R2": r2,
        "MSE": mse,
        "Predictors": predictors,
        "Error": error
    })

# --- Compile results ---
results_df = pd.DataFrame(results)
valid_results = results_df.dropna(subset=["R2", "MSE"])

best_r2_model = valid_results.loc[valid_results["R2"].idxmax()] if not valid_results.empty else None
best_mse_model = valid_results.loc[valid_results["MSE"].idxmin()] if not valid_results.empty else None

# --- Summary ---
print(f"\n✅ Valid regressions: {len(valid_results)} / {len(results_df)}")
print(f"❌ Failed regressions: {results_df['Error'].notna().sum()}")

if best_r2_model is not None:
    print("\n📈 Best R² model:")
    print(best_r2_model[["LagsIncluded", "R2", "MSE", "Predictors"]])

if best_mse_model is not None:
    print("\n📉 Best MSE model:")
    print(best_mse_model[["LagsIncluded", "R2", "MSE", "Predictors"]])
```

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd

# --- Settings ---
target_col = "x_202304"
base_predictors = ["imd_score_no_crime", "Population Density"]
num_lags = 20

# Sort lagged columns descending by date (most recent first)
all_lagged_cols = sorted(
    [col for col in df4.columns if col.startswith("x_") and col != target_col],
    key=lambda x: int(x.split("_")[1]),
    reverse=True
)

# Select predictors
selected_lags = all_lagged_cols[:num_lags]
predictors = base_predictors + selected_lags

# Clean and prepare data
X = df4[predictors].apply(pd.to_numeric, errors="coerce")
y = pd.to_numeric(df4[target_col], errors="coerce")

df_model = pd.concat([X, y.rename("y")], axis=1).dropna()
X_clean = df_model.drop(columns="y")
y_clean = df_model["y"]

# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y_clean, test_size=0.3, random_state=42
)

X_train_const = sm.add_constant(X_train)

# Poisson GLM to estimate dispersion (alpha)
poisson_model = sm.GLM(y_train, X_train_const, family=sm.families.Poisson()).fit()
pearson_chi2 = sum(((y_train - poisson_model.mu) ** 2) / poisson_model.mu)
alpha_est = pearson_chi2 / poisson_model.df_resid

# Negative Binomial GLM with estimated alpha
nb_family = sm.families.NegativeBinomial(alpha=alpha_est)
nb_model = sm.GLM(y_train, X_train_const, family=nb_family).fit()

# Predict on test set
X_test_const = sm.add_constant(X_test)
y_pred = nb_model.predict(X_test_const)

# Metrics
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# --- Output ---
print(nb_model.summary())
print(f"\nEstimated alpha: {alpha_est:.6f}")
print(f"Test set MSE: {mse:.6f}")
print(f"Test set R²: {r2:.6f}")
```
